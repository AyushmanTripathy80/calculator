#importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#importing Dataset

```python
dataset = pd.read_csv('heart.csv')
x = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
print(x)
```

```python
print(y)
```

#data preprocessing

#Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le1 = LabelEncoder()
le2 = LabelEncoder()
le6 = LabelEncoder()
le8 = LabelEncoder()
le10 = LabelEncoder()
x[:,1] = le1.fit_transform(x[:,1])
x[:,2] = le2.fit_transform(x[:,2])
x[:,6] = le6.fit_transform(x[:,6])
x[:,8] = le8.fit_transform(x[:,8])
x[:,10] = le10.fit_transform(x[:,10])
```

```python
print(x[418])
```

```python
print(y)
```

#Splitting Dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
print(X_train)
```

```python
print(Y_test)
```

#Feature scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
print(X_train)
```

#Training Dataset

```python
#logistic Regression
from sklearn.linear_model import LogisticRegression
model_logistic = LogisticRegression()
model_logistic.fit(X_train,Y_train)
```

```python
#svm
from sklearn.svm import SVC
model_svm = SVC()
model_svm.fit(X_train,Y_train)
```

```python
#Kneighbores
from sklearn.neighbors import KNeighborsClassifier
model_knn = KNeighborsClassifier(n_neighbors=7)
model_knn.fit(X_train,Y_train)
```

```python
#Decission Tree
from sklearn .tree import DecisionTreeClassifier
model_dt = DecisionTreeClassifier()
model_dt.fit(X_train,Y_train)
```

```python
#Random forest
from sklearn.ensemble import RandomForestClassifier
model_random = RandomForestClassifier(n_estimators=150)
model_random.fit(X_train,Y_train)
```

```python
#XGBoost
from xgboost import XGBClassifier
model_xg = XGBClassifier()
model_xg.fit(X_train,Y_train)
```

```python
y_pred_logistic = model_logistic.predict(X_test)
y_pred_svm = model_svm.predict(X_test)
y_pred_knn = model_knn.predict(X_test)
y_pred_dt = model_dt.predict(X_test)
y_pred_random = model_random.predict(X_test)
y_pred_xg = model_xg.predict(X_test)
```

#Testing accuracy

```python
from sklearn.metrics import accuracy_score
Logistic_Acc = accuracy_score(Y_test,y_pred_logistic)
SVM_Acc = accuracy_score(Y_test,y_pred_svm)
KNN_Acc = accuracy_score(Y_test,y_pred_knn)
DT_Acc = accuracy_score(Y_test,y_pred_dt)
Random_Acc = accuracy_score(Y_test,y_pred_random)
Xg_Acc = accuracy_score(Y_test,y_pred_xg)
```

```python
print(SVM_Acc)
print(Logistic_Acc)
print(KNN_Acc)
print(DT_Acc)
print(Random_Acc)
print(Xg_Acc)
```

#plotting Accuracy of different ML Algorithm

```python
plt.figure(figsize=(12,6))
plt.bar("Logistic Regression",Logistic_Acc,width=0.5)
plt.bar("KNeighbors",KNN_Acc,width=0.5)
plt.bar("support vector machine",SVM_Acc,width=0.5)
plt.bar("Decision Tree",DT_Acc,width=0.5)
plt.bar("Random Forest",Random_Acc,width=0.5)
plt.bar("XGBoost",Xg_Acc,width=0.5)
plt.xlabel('Machine Learning Algorithm')
plt.ylabel('Accuracy')
```
